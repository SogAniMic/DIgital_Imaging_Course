1- Qu'est ce qu'une image ?
   - Une image est représenté comme un tableau (ou matrice) de pixels 
   - Un pixel est tout simplement un point de l'image. Une image est donc un ensemble de point disposer sur 2 dimensions.
   - Regarder de plus prêt ou zoomer en grand l'une de vos photos numérique Vous verez apparaître des pixels ous forme de petit       carré
2- Les différentes type d'images : 
   Si chaque pixel trouve sa place logique dans une image, qu'en est il de la couleur et de l'intensité des pixels.
   Nous allons distinguer trois possibilités 
  a- L'image en noir et blanc
   - qui dit noir ou blanc dit binaire donc 0 ou 1. Donc dans le cas d'une image en noir et blanc, la valeur des pixels est          binaire (0 ou 1)
   - Une valeur de 0 représente un carré (pixel) noir 
   - Une valeur de 1 représente un carré (pixel) blanc
   - Une image en noir et blanc n'est donc qu'un échequier non régulier et sa matrice est une matrice de dimension 2 n'ayant que      des valeurs 0 et 1

## 1 - Création automatique d'image avec numpy

- Jusqu'ici, pour créer des images, nous mentionnons manuellement la valeur des pixels dans le tableau numpy correspondant

- On peut automatiser cette tâche en initialisant le tableau numpy, puis en lui affectant la valeur des pixels

## 1-1 Création d'image blanc noir
 
 Pour créer une image en noir et blanc ou en nuance de gris, on utilise un tableau de dimension 2 puis on modifie ses valeurs
 image = np.zeros([hauteur,largeur])
- haueur = nbre de pixels le long de la hauteur de l'image
- largeur = nbre de pixels le long de la largeur de l'image

```python
# Exemple de création d'une ilage en nuance de gris

# Etape 1 : initialisation du tableau

import numpy as np

image = np.zeros([2,12])

print(image)
```

```python
# Etape 2 : modification des valeurs
# modifier les valeurs du 1er sous tableau
image[0, 0:7] = 10
image[0,7:12] = 55
print(image)
```

```python
image[1, 0 :4] = 110 
image[1, 4:] = 70
print(image)
```

```python
# affichage de l'image

import matplotlib.pyplot as plt

imgplot = plt.imshow(image,cmap='Greys_r')
plt.show()
```

## 1-2 Création d'une image en couleur
Pour créer une image en couleurs, il suffit d'initialiser un tableau de dimension 3 et de modifier ses valeurs. Le tableau à utiliser est de la forme image = np.zeros([h,l,c])
h = nbre de pixels le long de la hauteur de l'image
l = nbre de pixes le long de la largeur de l'image
c = le canal de l'image (il vaut 3 car avec les couleurs, nous utilisons le format Rouge-Vert-Bleu (RVB))

```python
# Exemple de création d'une image en couleur
#Etape 1 : initialisation du tableau de pixels
image = np.zeros([4,3,3])
print(image)
```

```python
# Etape 2 : modification de la valeur des pixels
image[:,0,:] = (255,0,0)
print(image)
```

```python
image[:,1,:] = (0,255,0)
print(image)
```

```python
image[:,2,:] = (0,0,255)
print(image)
```

```python
# affichage de l'image

implot = plt.imshow(image)
```

# 2 - Lire une image avec PLL (Python Image Library)

```python
from PIL import Image
```

```python
# lecture du fichier image 

img = Image.open('test.jpg')
```

```python
# Obtention du tableau de pixel correspondant à l'image
img_array = np.array(img)

# affichage du tableau de pixels

print(img_array)
```

```python

implot = plt.imshow(img_array)
```

```python
# Dimensions du tableau de pixels obtenu
print('Dimensions: ', img_array.shape)
```

# 3 - Utilisation de la librairie PIL pour créer une image à partir d'un tableau de pixels

```python
pil_image = Image.fromarray(np.uint8(img_array))
pil_image
```

```python
# Enregistrer l'image dans le dossier courant
pil_image.save('mon_image.png')
```
